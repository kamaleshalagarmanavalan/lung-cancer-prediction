# Data Preprocessing

```python
data.isnull()
```

```python
data.isnull().sum() #checks unfilled columns and null values
```

```python
data=data.dropna() #deletes null values
```

```python
data.drop(['Name', 'Surname'], axis = 1)
```

## Feature Extraction

```python
y=data['Result']
x=data.drop(['Name','Surname' ,'Result'], axis=1) #axis 1 indicates colunmn, 0 - rows, drop processes other than that column
```

## Model creation

```python
from sklearn.model_selection import train_test_split
xtrain,xtest, ytrain, ytest=train_test_split(x,y,test_size=0.2, random_state=42) #0.2 = 75, 25// 0.3 = 70, 30
```

```python
xtrain.shape
```

```python
xtest.shape
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
from sklearn.metrics import accuracy_score  #accuracy_score: to evaluate the model's accuracy.
from sklearn.ensemble import RandomForestClassifier #RandomForestClassifier: a machine learning model that builds multiple decision trees and combines them to improve accuracy.
clf=RandomForestClassifier() #Random Forest classifier object named clf is created. This model will learn patterns in your training data.
clf.fit(xtrain,ytrain) #training the model using xtrain and ytrain
ypred=clf.predict(xtest) #to ask the trained model to predict outcomes for test data (xtest). The predictions are stored in ypred.
score=accuracy_score(ytest, ypred) #compare the predicted labels (ypred) with the actual labels (ytest) to compute the accuracy
print(score)

```

##Prediction

```python
k= clf.predict([[68,20,4,6]]) #values of the patient is given for the classification.
print(k)
```

```python
if (k==1):
  print("Lung cancer")
else:
  print("Not affected")
```
